```python
### Training with manually updating W with "Backward" ###

import torch
#from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F

import torch.optim as optim




data = [(1.0,2.1,3.0), (2.0, 3.5, 6.0), (3.0, 3.0, 9.0), (4.0, 2.1, 12.0), (5.0, 7.2, 15.0), (6.0, 10.1, 18.0)]


class Net(nn.Module): ## nn.Module class is used
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(2,3,bias=False)  # in dim, out dim
    def forward(self, x):
        x = self.fc1(x)
        return x

net = Net()

print(net)
print(list(net.parameters()))
```

```python


#input = torch.randn(1)
#out = net(input)

#def criterion(out, label):
#    return (label - out)**2
criterion = nn.MSELoss()


optimizer = optim.SGD(net.parameters(), lr=0.01, momentum=0.5)
#optimizer = optim.Adam(net.parameters(), lr=0.005)


for epoch in range(20): # 0 - 19
    for i, current_data in enumerate(data):
        X, Y, Z = current_data
        X = [X,Y]
        X, Z = torch.FloatTensor([X]), torch.FloatTensor([Z])
        optimizer.zero_grad()   
        outputs = net(X)
        loss = criterion(outputs, Z)
        loss.backward()
        optimizer.step()    ## This line is equivalent to "W = W - lr* W.grad"
        print("Epoch {} - loss: {}".format(epoch, loss))

### Test the trained network ###            
for i, current_data in enumerate(data):
    X, Y, Z = current_data
    X = [X,Y]
    X, Z = torch.FloatTensor([X]), torch.FloatTensor([Z])
    out = net(torch.FloatTensor(X))  
    print("when x, y = {}, Z = {}".format(X, out))
    
```
